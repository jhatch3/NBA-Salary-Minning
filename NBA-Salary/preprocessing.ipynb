```python
import pandas as pd 
import numpy as np 
```

```python
years = ["2020-21", "2021-22", "2022-23", "2023-24"]

nba_csv_files = [
    "wizards.csv",
    "warriors.csv",
    "timberwolves.csv",
    "thunder.csv",
    "suns.csv",
    "spurs.csv",
    "rockets.csv",
    "raptors.csv",
    "pistons.csv",
    "pelicans.csv",
    "pacers.csv",
    "nuggets.csv",
    "nets.csv",
    "mavs.csv",
    "magic.csv",
    "lakers.csv",
    "knicks.csv",
    "kings.csv",
    "jazz.csv",
    "hornets.csv",
    "heat.csv",
    "hawks.csv",
    "grizzlies.csv",
    "clippers.csv",
    "celtics.csv",
    "cavs.csv",
    "bulls.csv",
    "bucks.csv",
    "blazers.csv",
    "76ers.csv",
]

# Max sure we have all 30 teams 
len(nba_csv_files) == 30
```

```python
for year in years: 
    for team in nba_csv_files:
        file_path = ("Data/dirty/" + str(year) + "/" + str(team))
        df = pd.read_csv(file_path)cl
        mp = df["MP"].astype(int).sum()
        print(f"{year} - {team.replace('.csv','')}: {mp} mp")
         
     

```

```python
key = "Player-additional" 
df["SeasonType"] = df.groupby(key).cumcount().map({0: "RS", 1: "PO"})


id_cols = [key, "Player"]
value_cols = [c for c in df.columns if c not in id_cols + ["SeasonType"]]

wide = (
    df.pivot(index=id_cols, columns="SeasonType", values=value_cols)
      .sort_index(axis=1)
)

# Flatten multiindex columns -> "PTS_RS", "PTS_PO", etc.
wide.columns = [f"{stat}_{season}" for stat, season in wide.columns]
wide = wide.reset_index()

wide["team"] = "cavaliers"
wide
```

```python
wide["salary"] = [
    20000000,
    559782,
    36725670,
    2300000,
    21696429,
    2560975,
    16615384,
    2164993,
    35410310,
    11227657,
    8500000,
    float('nan'),
    10185186,
    1890000,
    15212068,
    3000000,
    559782,
    float('nan'),
    559782,
    float('nan'),
    6166667
]

wide = wide[wide["salary"].notna()] 

# 1) Turn obvious "missing" strings into real NaN everywhere
wide = wide.replace(
    to_replace=[r'^\s*$', r'^\s*NaN\s*$', r'^\s*nan\s*$', r'^\s*None\s*$', r'^\s*null\s*$'],
    value=np.nan,
    regex=True
)

# 2) For each column, try numeric conversion; if it produces any numbers, keep numeric
for col in wide.columns:
    s_num = pd.to_numeric(wide[col], errors="coerce")
    # if at least one value successfully became a number (non-NaN), keep it numeric
    if s_num.notna().any():
        wide[col] = s_num
    else:
        wide[col] = wide[col].astype("string")  # keep as proper string dtype

wide 
```
