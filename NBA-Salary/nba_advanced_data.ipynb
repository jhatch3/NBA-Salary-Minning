```python
# Only need name, pos, mp, OBPM, DBPM, VORP, Player-additional
def process_columns(df, team:str, year:str) -> pd.DataFrame:

    # Drop any columns we dont need
    df = df.drop(columns=['Rk', 'Age', 'Pos', 'G', 'GS', 'PER', 'TS%', '3PAr',
        'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%',
        'OWS', 'DWS', 'WS', 'WS/48', 'BPM', 'Awards'], axis=1)

    # add team and year to df
    df["Team"] = team
    df["year"] = year 

    return df 

```

```python

tmp_df = pd.DataFrame() 

for year in years: 
    for team in nba_csv_files:
        file_path = ("Data/dirty/" + str(year) + "/" + str(team))
        df = pd.read_csv(file_path)
        df = process_columns(df, file_to_tm[str(team)], int('20' + year.split("-")[1]))
        tmp_df = pd.concat([tmp_df, df])
        df.to_csv("Data/clean/" + str(year) + "/" + str(team), index=False)
    break 

tmp_df.to_csv("Data/clean/nba_advanced_data.csv", index=False)
tmp_df
```
